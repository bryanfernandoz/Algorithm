A contour plot represents a 3D surface by plotting lines that connect points with common Z-values along a slice.
E.g.: Visualize a plane in 2D and 3D


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x,y):
  return np.sin(x)**2+np.cos(y)**2
```

## Generate 50 values from 0 to 5

```python
x=np.linspace(0,5,50)
y=np.linspace(0,5,50)
```

## generate combinations of grids

```python
x,y=np.meshgrid(x,y)
z=func(x,y)
```

## Draw plot

```python
plt.contour(x,y,z,cmap='gist_rainbow_r');
plt.show()
```
